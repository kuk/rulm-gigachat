```python
%run -n main.py

messages = []
limit = 50
for offset in log_progress(range(0, 1000, limit)):
    data = gigachat_session_messages(SESSION_ID, offset=offset, limit=limit)
    chunk = list(parse_session_messages(data))
    if not chunk:
        break

    messages.extend(chunk)
len(messages)
```

```python
%run -n main.py
prompt_answers = dict(message_prompt_answers(messages))
len(prompt_answers)
```

```python
for item in log_progress(items):
    answer = item.get('answer')
    if answer:
        continue
        
    prompt = item_prompt(item)
    answer = prompt_answers.get(prompt)
    if answer:
        item['answer'] = answer
        continue

    print(item)
    print()

    data = gigachat_request(prompt, session_id=SESSION_ID)
    request_id = parse_gigachat_request(data)
    lines = gigachat_result_events(request_id, space_id=SPACE_ID, user_id=USER_ID)
    events = parse_gigachat_result_events(lines)
    answer = ready_event_text(events)
    assert answer
    item['answer'] = answer
    
    print(answer)
    print('\n\n')
```
