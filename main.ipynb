```python
%run -n main.py
lines = read_lines('data/user_oriented_ru_v2.jsonl')
items = list(parse_jsonl(lines))
```

```python
for item in log_progress(items):
    answer = item.get('answer')
    if answer:
        continue
        
    prompt = item_prompt(item)
    answer = prompt_answers.get(prompt)
    if answer:
        item['answer'] = answer
        continue

    print(item)
    print()

    data = gigachat_request(prompt, session_id=SESSION_ID)
    request_id = parse_gigachat_request(data)
    lines = gigachat_result_events(request_id, space_id=SPACE_ID, user_id=USER_ID)
    events = parse_gigachat_result_events(lines)
    answer = ready_event_text(events)
    assert answer
    item['answer'] = answer
    
    print(answer)
    print('\n\n')
```
